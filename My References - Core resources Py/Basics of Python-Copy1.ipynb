### LOOPS
* A loop is a programming structure that executes the same code over and over again, as long as a certain condition is met

#### WHILE LOOP
* A while loop executes the code in its block while a condition is met.

```python
number = 0
while number < 10 :
    number += 1
    print (number)
```

#### FOR LOOP
* The for loop works a bit differently. Here we dont have a condition. This loop type is used to iterate over sequences

```python
numbers = [ 10 , 20 , 30 , 40 ]
for number in numbers:
    print (number)
```

```python
numbers = [ 10 , 20 , 30 , 40 ]
for x in numbers:
    print (x)
```

```python
players = ["Manu", "Anu", "Vinu"]

for player in players:
    print(player)
```

### RANGE FUNCTION
* With the range function, we can create lists that contain all numbers in between two numbers.

```python
for x in range ( 100 ):
    print (x)
```

```python
for i in range(20):
    print(i)
```

```python
for x in range ( 20 , 80 ):
    print (x)
```

```python
for x in range ( 1 , 100, 2):
    print (x)
```

### LOOP CONTROL STATEMENTS
*In order to manage loops, we have so-called loop control statements . They allow us to manipulate the process flow of the loop at a specific point.

#### BREAK STATEMENT
* With the break statement, we can end a loop immediately, without caring about the condition.

```python
number = 0
while number < 10 :
    number += 1
    if number == 5 :
        break
    print (number)
```

#### CONTINUE STATEMENT
* If we dont want to break the full loop, but to skip only one iteration, we can use the continue statement.

```python
number = 0
while number < 10 :
    number += 1
    if number == 4 :
        continue
    print (number)
```

#### PASS STATEMENT
* The pass statement is a very special statement, since it does absolutely nothing. Actually, it is not really a loop control statement, but a placeholder for code.
* Sometimes you want to write your basic code structure, without implementing the logic yet. In this case, we can use the pass statement, in order to fill the code blocks. Otherwise, we cant run the script.

```python
if number == 10 :
    pass
else :
    pass
while number < 10 :
    pass
```
