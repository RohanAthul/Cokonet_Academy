# Plotting with pandas

```python
import pandas as pd 
import numpy as np 
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('./Car_sales.csv', parse_dates=['Latest_Launch'])
data.head(10)
```

```python
data.info()
```

```python
data.describe().T
```

```python
data.isnull().mean()*100
```

```python
data.dropna(inplace=True)
data.drop_duplicates(inplace=True)
```

```python
data.shape
```

## Box plot

```python
data.Length.plot.box();
```

```python
data.Horsepower.plot.box(cmap = 'seismic');
```

## Kernel Density Estimation Plot (KDE PLOT)

```python
data.Length.plot.kde();
```

```python
data.Sales_in_thousands.plot.kde(cmap = 'flag');
```

### Sub plots 

```python
data.plot.box(subplots = True);
```

```python
data.plot.box( figsize=(18,3), subplots = True);
```

```python
data.plot.hist( figsize=(6,20), subplots = True);
```

```python
data.plot.kde(figsize = (5,20),subplots = True);
```

## Line charts

```python
data.plot.line(x='__year_resale_value', y='Price_in_thousands', cmap = 'Spectral');
```

#### This happens due to the fact that year is not in order.

```python
data.sort_values(by='__year_resale_value').plot.line(x='__year_resale_value', y='Price_in_thousands', cmap='Wistia');
```

```python
data.plot.bar(x='Manufacturer', y='Sales_in_thousands', figsize = (20, 5),cmap = 'gist_earth');
```

### Count plot

```python
data.Manufacturer.value_counts()
```

```python
data.Manufacturer.value_counts().plot.bar();
```

```python
data.Vehicle_type.value_counts().plot.bar(figsize=(12,5), cmap = 'Paired');
```

## Area chart

```python
data.plot.area(x='Manufacturer', y='Sales_in_thousands', figsize = (10,6), cmap = 'rainbow');
```

## Horizontal bar chart

```python
data.nlargest(5, 'Price_in_thousands').plot.barh(x = 'Model', y = 'Price_in_thousands');
```

```python
data.sort_values(by='Price_in_thousands')[:5].plot.barh(x='Model', y= 'Power_perf_factor', figsize = (6,4), cmap = 'flag');
```

## Scatter plot

```python
data.plot.scatter(x = 'Engine_size', y = 'Horsepower');
```

### Pie chart

```python
data.Vehicle_type.value_counts()
```

```python
data.Vehicle_type.value_counts().plot.pie();
```

```python
data.Vehicle_type.value_counts().plot.pie(autopct ='%1.2f%%', legend = True);
```

```python
data.Vehicle_type.value_counts().plot.pie(autopct ='%1.0f%%', legend = True);
```

```python
data.Vehicle_type.value_counts().plot.pie(autopct ='%1.4f%%', legend = True);
```

```python
data.plot.hexbin('Curb_weight', 'Fuel_capacity', gridsize=25);
```
